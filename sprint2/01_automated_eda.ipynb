# 2.1 Automated Exploratory Data Analysis (EDA)

Generate comprehensive EDA reports with minimal code using ydata-profiling, Sweetviz, and Dataprep.

## Learning Objectives
- Generate automated HTML profiling reports
- Compare train/test splits visually
- Interpret AI-generated insights

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

# Load sample dataset
data = load_breast_cancer(as_frame=True)
df = data.frame
df['target'] = data.target

train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
print(f'Train shape: {train_df.shape}, Test shape: {test_df.shape}')
```

```python
from ydata_profiling import ProfileReport

# Generate a full profiling report
profile = ProfileReport(train_df, title='Breast Cancer Dataset Profile', explorative=True)
profile.to_file('sprint2_eda_report.html')
print('Report saved to sprint2_eda_report.html')
```

```python
import sweetviz as sv

# Compare train and test distributions
report = sv.compare([train_df, 'Train'], [test_df, 'Test'], target_feat='target')
report.show_html('sweetviz_comparison.html')
print('Sweetviz comparison saved.')
```

```python
# LLM-powered EDA insight generation
from openai import OpenAI
import json

client = OpenAI()

def llm_eda_insights(df_stats: dict) -> str:
    """Use an LLM to generate EDA insights from summary statistics."""
    stats_str = json.dumps(df_stats, indent=2, default=str)
    response = client.chat.completions.create(
        model='gpt-4',
        messages=[{
            'role': 'user',
            'content': f'Analyse these dataset statistics and provide 5 key insights:\n{stats_str}'
        }]
    )
    return response.choices[0].message.content

stats = df.describe().to_dict()
# insights = llm_eda_insights(stats)
# print(insights)
```
