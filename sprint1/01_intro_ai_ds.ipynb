```python
import openai

# Example: Ask an LLM for data science guidance
# Make sure to set your API key: export OPENAI_API_KEY='your-key'
response = openai.ChatCompletion.create(
    model='gpt-4-turbo',
    messages=[
        {'role': 'system', 'content': 'You are a data science expert.'},
        {'role': 'user', 'content': 'Describe what kind of analysis I should perform on a customer churn dataset.'}
    ]
)
print(response.choices[0].message.content)
```

## Key Takeaways
- LLMs can generate, explain, and optimise data science code
- Agentic workflows combine search, code execution, and reasoning
- Always consider bias, transparency, and privacy when using AI tools
