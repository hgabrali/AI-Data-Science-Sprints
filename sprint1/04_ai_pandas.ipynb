```python
import pandas as pd
from pandasai import SmartDataframe

# Create a sample DataFrame
data = {
    'region': ['North', 'South', 'East', 'West', 'North', 'South'],
    'product': ['A', 'B', 'A', 'C', 'B', 'C'],
    'sales': [1500, 2300, 1800, 900, 2100, 1600],
    'units': [150, 230, 180, 90, 210, 160]
}
df = pd.DataFrame(data)
print(df.head())
```

```python
from openai import OpenAI

client = OpenAI()

def pandas_code_gen(question: str, df_info: str) -> str:
    """Generate pandas code from a natural language question."""
    prompt = f'Given a DataFrame with columns: {df_info}\nWrite pandas code to: {question}\nReturn only the code.'
    response = client.chat.completions.create(
        model='gpt-4',
        messages=[{'role': 'user', 'content': prompt}]
    )
    return response.choices[0].message.content

# df_info = str(df.dtypes.to_dict())
# code = pandas_code_gen('Find the top 3 regions by total sales', df_info)
# print(code)
```
