# 1.2 AI-Based Web Search as the First Step

Learn how to combine LLMs with search engines for automated data discovery.


```python
# !pip install langchain langchain-community duckduckgo-search
```

```python
from langchain.tools import DuckDuckGoSearchRun

search = DuckDuckGoSearchRun()
results = search.run('best public datasets for machine learning 2024')
print(results)
```

```python
from langchain.agents import Tool, create_react_agent, AgentExecutor
from langchain_openai import ChatOpenAI
from langchain import hub

llm = ChatOpenAI(model='gpt-4-turbo', temperature=0)
tools = [Tool(name='Search', func=search.run, description='Search the web for information')]
prompt = hub.pull('hwchase17/react')

agent = create_react_agent(llm, tools, prompt)
executor = AgentExecutor(agent=agent, tools=tools, verbose=True)

# Ask the agent to find a dataset
result = executor.invoke({'input': 'Find a public CSV dataset about customer churn suitable for machine learning'})
print(result['output'])
```
