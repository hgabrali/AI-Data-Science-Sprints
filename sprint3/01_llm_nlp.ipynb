## 1. Sentiment Analysis


```python
from openai import OpenAI

client = OpenAI()

def analyze_sentiment(text: str) -> str:
    response = client.chat.completions.create(
        model='gpt-3.5-turbo',
        messages=[
            {'role': 'system', 'content': "Classify the sentiment. Output only 'positive', 'negative', or 'neutral'."},
            {'role': 'user', 'content': text}
        ],
        max_tokens=10
    )
    return response.choices[0].message.content.strip().lower()

# texts = [
#     'The product quality is excellent, but shipping was slow.',
#     'Worst experience ever.',
#     'It arrived on time and works as described.'
# ]
# for t in texts:
#     print(f'{analyze_sentiment(t):10} | {t}')
```

## 2. Named Entity Recognition with Structured Output


```python
import json

def extract_entities(text: str) -> dict:
    response = client.chat.completions.create(
        model='gpt-4',
        response_format={'type': 'json_object'},
        messages=[{
            'role': 'user',
            'content': f'Extract named entities. Return JSON with keys: persons, organizations, locations, dates.\n\nText: {text}'
        }]
    )
    return json.loads(response.choices[0].message.content)

# text = 'Apple CEO Tim Cook announced in San Francisco on January 15, 2024 that Vision Pro launches globally.'
# print(json.dumps(extract_entities(text), indent=2))
```

## 4. Text-to-SQL


```python
def text_to_sql(question: str, schema: str) -> str:
    response = client.chat.completions.create(
        model='gpt-4',
        messages=[{'role': 'user', 'content': f'Schema:\n{schema}\n\nQuestion: {question}\n\nWrite a SQL query. Return only the SQL.'}]
    )
    return response.choices[0].message.content

# schema = 'Table: orders (order_id, customer_id, amount, order_date)'
# print(text_to_sql('Top 5 customers by total spend last 30 days', schema))
```
